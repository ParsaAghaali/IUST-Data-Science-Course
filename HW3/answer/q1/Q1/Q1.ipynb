1.2 - Read the data

* **credit.policy**: 1 if the customer meets the credit underwriting criteria of LendingClub.com, and 0 otherwise.
* **purpose**: The purpose of the loan (takes values `credit_card`, `debt_consolidation`, `educational`, `major_purchase`, `small_business`, and `all_other`).
* **int.rate**: The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers judged by LendingClub.com to be more risky are assigned higher interest rates.
* **installment**: The monthly installments owed by the borrower if the loan is funded.
* **log.annual.inc**: The natural log of the self-reported annual income of the borrower.
* **dti**: The debt-to-income ratio of the borrower (amount of debt divided by annual income).
* **fico**: The FICO credit score of the borrower.
* **days.with.cr.line**: The number of days the borrower has had a credit line.
* **revol.bal**: The borrower's revolving balance (amount unpaid at the end of the credit card billing cycle).
* **revol.util**: The borrower's revolving line utilization rate (the amount of the credit line used relative to total credit available).
* **inq.last.6mths**: The borrower's number of inquiries by creditors in the last 6 months.
* **delinq.2yrs**: The number of times the borrower had been 30+ days past due on a payment in the past 2 years.
* **pub.rec**: The borrower's number of derogatory public records (bankruptcy filings, tax liens, or judgments).

3.1 - Why is it necessary to convert categorical data into numeric data for machine learning models? How does this process affect the model's performance? Should all categorical data be converted to numeric data, and does this conversion always have a positive impact on the model's performance?

### Answer

Converting categorical data into numeric data is essential for machine learning models because most models work with numerical inputs and cannot directly process categorical variables. Here's why and how it impacts performance:

Why Convert Categorical Data?
Mathematical Operations: Machine learning models often rely on mathematical operations, such as matrix multiplication, which require numerical inputs.
Representation of Information: Numeric conversion allows categorical information to be encoded in a way that the model can interpret relationships between data points.
Consistency Across Models: Some models (e.g., decision trees) handle categorical data inherently, but others (e.g., logistic regression, neural networks) require numerical input.
How It Affects Model Performance:
Positive Impact:

Proper encoding (e.g., one-hot encoding or label encoding) ensures the model understands categorical relationships.
Prevents information loss by maintaining feature importance when encoded correctly.
Improves generalization by allowing models to better capture patterns.
Negative Impact:

Poor encoding (e.g., arbitrarily assigning numbers) can introduce unintended ordinal relationships where none exist, leading to biased results.
One-hot encoding of high-cardinality categorical variables can increase dimensionality, causing overfitting or computational inefficiency.
Should All Categorical Data Be Converted?
Yes, categorical data should be converted into numeric form for most machine learning models. However:

Careful Encoding: The choice of encoding (e.g., one-hot, label encoding, target encoding) should match the nature of the data and the model.
Avoid Over-Transformation: If a model inherently supports categorical data (e.g., tree-based models like Random Forest), conversion may not always be necessary.
Does Conversion Always Have a Positive Impact?
No, conversion does not always have a positive impact. It depends on:

Choice of Encoding: Incorrect encoding can misrepresent the data, leading to poor performance.
Cardinality: High-cardinality features can lead to sparsity in the dataset, causing inefficiencies.
Model Requirements: For some models, inappropriate encoding can lead to overfitting or underfitting.
Conclusion:
While converting categorical data is often necessary, it should be done thoughtfully with an understanding of the data, the encoding method, and the machine learning model to ensure it positively impacts performance.








# 4. Implementing model (Decision Tree & Random Forest)

4.1 - Train a Decision Tree model on the training dataset

```python
X_train = train_data.drop(columns=["not.fully.paid"])
y_train = train_data["not.fully.paid"]

X_validation = validation_data.drop(columns=["not.fully.paid"])
y_validation = validation_data["not.fully.paid"]
```

```python
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
```

```python
random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, y_train)
```

# 5. Prediction & Evaluation

5.1 - Please make a prediction for the test dataset

```python
decision_tree_preds = decision_tree.predict(X_validation)
decision_tree_probs = decision_tree.predict_proba(X_validation)[:, 1]
```

```python
random_forest_preds = random_forest.predict(X_validation)
random_forest_probs = random_forest.predict_proba(X_validation)[:, 1]
```

5.2 - For the evaluation of classification models, the following metrics are commonly used: **Precision**, **Recall**, **F1-Score (Macro)**, **F1-Score (Micro)**, and **ROC-AUC**.  

- **a)** Define each metric and explain how it is calculated.  
- **b)** Discuss what each metric represents in terms of model performance and in which scenarios it is most useful.  
- **c)** Compare **F1-Score (Macro)** and **F1-Score (Micro)**, highlighting the differences in their calculations and applications.  
- **d)** Explain how the **ROC-AUC** metric helps in evaluating model performance across different classification thresholds.  

Make sure to provide examples or scenarios where each metric would be particularly valuable.

```python
def evaluate_model(name, y_true, y_pred, y_prob):
    print(f"\n===== Metrics for {name} =====\n")
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred))
    print(f"F1-Score (Macro): {f1_score(y_true, y_pred, average='macro'):.4f}")
    print(f"F1-Score (Micro): {f1_score(y_true, y_pred, average='micro'):.4f}")
    if y_prob is not None:
        print(f"ROC-AUC: {roc_auc_score(y_true, y_prob):.4f}")

evaluate_model("Decision Tree", y_validation, decision_tree_preds, decision_tree_probs)
evaluate_model("Random Forest", y_validation, random_forest_preds, random_forest_probs)
```
