Exercise 7:<br/>
Read the total profit of each month and show it using the histogram to see the most common profit ranges<br/>
The histogram should look like this.<br/>




```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('sales.csv')

total_sales = {
    'Face Cream': data['facecream'].sum(),
    'Face Wash': data['facewash'].sum(),
    'Toothpaste': data['toothpaste'].sum(),
    'Bathing Soap': data['bathingsoap'].sum(),
    'Shampoo': data['shampoo'].sum(),
    'Moisturizer': data['moisturizer'].sum()
}
products = list(total_sales.keys())
units_sold = list(total_sales.values())
plt.figure(figsize=(20, 15))  
wedges, texts, autotexts = plt.pie(units_sold, labels=products, autopct='%1.1f%%', startangle=0, radius=0.65, counterclock=True)  # Create pie chart with product labels

for text in texts:
    text.set_fontsize(16)  

for autotext in autotexts:
    autotext.set_fontsize(20)  

plt.legend(wedges, products, title="Products", loc="lower right", fontsize=20)
plt.title('Sales Data', fontsize=20)  
plt.gcf().patch.set_facecolor('white')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('sales.csv')

months = data['month_number']
bathing_soap_sales = data['bathingsoap']
facewash_sales = data['facewash']

fig, axs = plt.subplots(2, 1, figsize=(14, 10))  

axs[0].plot(months, bathing_soap_sales, marker='o', markersize=12, color='black', label='Bathing Soap', linewidth=6)
axs[0].set_title('Sales data of a Bathingsoap ', fontsize=16)
axs[0].set_xlabel('Month Number', fontsize=20)
axs[0].set_ylabel('sales units in number', fontsize=20)
axs[0].set_xticks(range(1, 13))  
axs[0].set_yticks([7500, 10000, 12500])  
axs[0].tick_params(axis='y', labelsize=20)  
axs[0].tick_params(axis='x', labelsize=20) 

axs[1].plot(months, facewash_sales, marker='o', color='red', markersize=12, label='Face Wash', linewidth=6)
axs[1].set_title('Sales data of a facewash', fontsize=20)
axs[1].set_xlabel('Month Number', fontsize=20)
axs[1].set_ylabel('sales units in number', fontsize=20)
axs[1].set_xticks(range(1, 13))  
axs[1].set_yticks([1500, 2000]) 
axs[1].tick_params(axis='y', labelsize=20)  
axs[1].tick_params(axis='x', labelsize=20) 

plt.tight_layout()
plt.show()
```

Exercise 9:<br/>
Read Bathing soap facewash of all months and display it using the Subplot<br/>
The Subplot should look like this.<br/>




```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the sales data from CSV file
data = pd.read_csv('sales.csv')
ww=[2,4,6,8,10,12]
# Extract month numbers and product sales data
months = data['month_number']
facecream_sales = data['facecream']
facewash_sales = data['facewash']
toothpaste_sales = data['toothpaste']
bathingsoap_sales = data['bathingsoap']
shampoo_sales = data['shampoo']
moisturizer_sales = data['moisturizer']

# Prepare data for stack plot
sales_data = [facecream_sales, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales]

# Define colors for each product
colors = ['purple', 'skyblue', 'red', 'black', 'green', 'yellow']

# Create stack plot with specified colors
plt.figure(figsize=(12, 8))
plt.stackplot(months, sales_data, labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'], colors=colors)

# Add titles and labels
plt.title('All product sale data using stack plot', fontsize=20)
plt.xlabel('Month Number', fontsize=20)
plt.ylabel('Units Sold', fontsize=20)
plt.xticks(ww, fontsize=20)  # Set x-axis ticks to month numbers
plt.yticks(fontsize=20)
plt.legend(loc='upper left', fontsize=20)  # Show legend
plt.gca().spines['left'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['top'].set_linewidth(2)
plt.tight_layout()
plt.show()
```
